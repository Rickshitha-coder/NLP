```python
from collections import defaultdict

corpus = [
    "best places to visit in india",
    "best places to visit in chennai",
    "best places to visit near me",
    "best places to visit during summer",
    "best places to visit in kerala",
    "best places to visit in delhi",
    "best places to visit with family"
]

D = 0.75

# Store 5-gram counts
fivegram = defaultdict(int)

# Store 4-word history counts
history = defaultdict(int)

for sentence in corpus:
    words = sentence.lower().split()

    for i in range(len(words)-4):
        gram = tuple(words[i:i+5])
        context = tuple(words[i:i+4])

        fivegram[gram] += 1
        history[context] += 1

# Continuation (previous contexts)
continuation = defaultdict(set)

for gram in fivegram:
    context = gram[:4]
    word = gram[4]
    continuation[word].add(context)

# Followers (next words)
followers = defaultdict(set)

for gram in fivegram:
    context = gram[:4]
    word = gram[4]
    followers[context].add(word)

total_unique_fivegrams = len(fivegram)

def kneser_ney(context, word):

    gram = context + (word,)

    count = fivegram.get(gram, 0)

    history_count = history.get(context, 0)

    if history_count == 0:
        return len(continuation[word]) / total_unique_fivegrams

    first = max(count - D, 0) / history_count

    lamb = (D * len(followers[context])) / history_count

    cont_prob = len(continuation[word]) / total_unique_fivegrams

    return first + lamb * cont_prob


# Query entered by user
query = "best places to visit"

context = tuple(query.lower().split()[-4:])

suggestions = [
    "india",
    "chennai",
    "kerala",
    "delhi",
    "me",
    "summer",
    "family"
]

print("Suggestions:\n")

for word in suggestions:
    print(word, ":", round(kneser_ney(context, word), 3))
```
