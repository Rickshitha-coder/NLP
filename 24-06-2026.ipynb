```python
import nltk
import string
import matplotlib.pyplot as plt
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
text=input()
#convert to lowercase
text=text.lower()
#remove punctuations
for t in string.punctuation:
    text=text.replace(t,"")

#tokenization
words=word_tokenize(text)
print("Tokens  : \n",words)

#total words
total_words=len(words)
print("Total Words : \n",total_words)

#vocabulary size - to find unqiue words
unique_words = set(words)
print("Vocabulary Size : \n",len(unique_words))

#word frequency 
freq = Counter(words)
print("\nWord Frequencies : \n",freq)

#top most frequent words
print("\n Top 10 words ")
for word, count in freq.most_common(10):
    print(word ," : ",count)


hapax_legomena=[]
for word,count in freq.items():
    if count == 1:
        hapax_legomena.append(word)
print("\n Rare Words : ",hapax_legomena)

#type token ratio (ttr)
ttr=len(unique_words) / total_words
print("\n Type Token Ratio : ",round(ttr,2))


#stop word removal
stop_words = set(stopwords.words('english'))
filtered_words=[ w for w in words if w not in stop_words]
print("\nfiltered Words : " ,filtered_words)

#word frequency after stop word removal
filtered_freq = Counter(filtered_words)
print("\nWord Frequencies : \n",filtered_freq)

#chat visulaisation
word_list=list(filtered_freq.keys())
counts_list=list(filtered_freq.values())
plt.figure(figsize=(8,5))
plt.bar(word_list, counts_list)
plt.title("NLP WORD FREQUECY ANALYSIS")
plt.xlabel("Words")
plt.ylabel("frequency")
plt.xticks(rotation=90)
plt.show()
```

```python
#Next word prediction using N- gram

from collections import defaultdict,Counter
#training corpus
text="""
What is AI
What is GenAI
What is LLM
What is RAG
What is Tokenization
What is LLM
"""
#tokenization
words=text.lower().split()

#build bigram
model=defaultdict(list)

for i in range(len(words)-2):
    w1,w2,w3= words[i],words[i+1],words[i+2]
    model[(w1,w2)].append(w3)

#Inout phrase
input_words=("what","is")

#predict
if input_words in model:
    prediction = Counter(model[input_words]).most_common(1)[0][0]
    print("Input : What is ")
    print("Predicted next word:", prediction)
else:
    print("Phrase not found")
```
