```python
""" Task 2
Scenario: Smart News Analysis System 
A media company receives thousands of news articles every day. To help journalists quickly 
understand the content, the company wants to build a Smart News Analysis System using NLTK. 
The system should analyze the news article and perform various NLP tasks using different NLTK 
modules. 
Tasks 
1. Parse the sentence structure to identify grammatical relationships using nltk.parse.  
2. Extract named entities such as persons, organizations, and locations using nltk.chunk.  
3. Classify the news article into categories such as Sports, Politics, Business, or Technology using 
nltk.classify.  
4. Evaluate the classification model using accuracy and precision measures with nltk.metrics.  
5. Calculate word frequencies and probability distributions of important terms using 
nltk.probability.  
6. Analyze semantic meaning and relationships between important words using nltk.sem.  
7. Generate bigrams and trigrams from the article using nltk.util. 
"""

import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import pos_tag, ne_chunk, FreqDist
from nltk.chunk import tree2conlltags
from nltk.classify import NaiveBayesClassifier, accuracy
from nltk.metrics import precision
from nltk.util import bigrams, trigrams
from nltk.probability import FreqDist

news = """
Microsoft announced a new AI platform in Chennai.
The company expects the technology to improve business productivity.
"""

print("Parsing:")

tokens = word_tokenize(news)
tagged = pos_tag(tokens)

verb_found = False

for word, tag in tagged:

    if tag.startswith('NNP'):
        print("Subject:", word)
        continue

    if tag.startswith('VB'):
        print("Verb:", word)
        verb_found = True
        continue

    if verb_found and tag.startswith('NN'):
        print("Object:", word)
        break

print('\nNamed Entities : ')

entities=ne_chunk(tagged)
for chunk in entities:
    if hasattr(chunk, 'label'):
        entity_name = " ".join(c[0] for c in chunk)
        print(chunk.label(), ":", entity_name)


print("\nCategory : ")

def features(text):
    words=word_tokenize(text.lower())
    return {word: True for word in words}
    
train_data = [
    (features("AI technology software computer"), "Technology"),
    (features("football cricket sports game"), "Sports"),
    (features("government election politics"), "Politics"),
    (features("business market finance company"), "Business")
]
classifier = NaiveBayesClassifier.train(train_data)

category = classifier.classify(features(news))
print(category)


test_data = train_data

acc = accuracy(classifier, test_data)

print("\nModel Accuracy:")
print(f"{acc*100:.0f}%")


refset = {'Technology'}
testset = {category}

if len(testset) > 0:
    p = precision(refset, testset)
    print("Precision:", p)


print("\nWord Frequency:")

words = [w.lower() for w in tokens if w.isalpha()]

fdist = FreqDist(words)

for word, freq in fdist.most_common():
    print(word, ":", freq)

print("\nProbability Distribution:")

total_words = len(words)

for word, freq in fdist.items():
    print(f"P({word}) = {freq}/{total_words} = {freq/total_words:.2f}")


print("\nSemantic Analysis:")

important_words = ["AI", "technology", "business"]

for word in important_words:
    print(word, "is related to innovation and productivity")


print("\nBigrams:")

token_words = [w for w in words]

for bg in bigrams(token_words):
    print(bg)

print("\nTrigrams:")

for tg in trigrams(token_words):
    print(tg)

```
