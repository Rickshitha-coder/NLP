```python
import nltk
import string
import matplotlib.pyplot as plt
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
text=input()
text=text.lower()
print("Lower case words : ",text)

#lowercase
for t in string.punctuation:
    text=text.replace(t,"")
#tokeniztion
words=word_tokenize(text)
print("Tokens  : \n",words)

#frequency 
freq=Counter(words)
print("\nWord Frequencies : \n",freq)


#vocabulary
unique_words=set(words)
print("Vocabulary Size : \n",len(unique_words))

#top most 5 words
print("\ Top Most 5 Words")
for word,count in freq.most_common(5):
    print(word, " : ",count)

#type token ratio
total_words=len(words)
ttr=len(unique_words)/total_words
print("\n Type Token Ratio : ",round(ttr,2))

#stop word removal
stop_words = set(stopwords.words('english'))
filtered_words=[ w for w in words if w not in stop_words]
print("\nfiltered Words : " ,filtered_words)

# Dominant Topic Analysis
filtered_freq = Counter(filtered_words)

print("\nDominant Topic Analysis")
topic, count = filtered_freq.most_common(1)[0]
print("Most Dominant Topic Keyword:", topic)
print("Frequency:", count)


#visulaization
word_list=list(filtered_freq.keys())
counts_list=list(filtered_freq.values())
plt.figure(figsize=(8,5))
plt.bar(word_list,counts_list)
plt.title("NLP FREQUENCY ANALYSIS")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=90)
plt.show()
```
