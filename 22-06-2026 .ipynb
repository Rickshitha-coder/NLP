```python
#Punctuation Removal
import string

text = 'Hello, World!'

result = text.translate(str.maketrans('', '', string.punctuation))
print(result)
```

```python
#Number Removal
import re
text="Pythonn 3.12 released in 2025"
result = re.sub(r'\d+', '', text)
print(result)
```

```python
#Regular Expressions (Regex)
import re
text = "Contact us at support@gmail.com"
emails = re.findall(r'\S+@\S+', text)
print(emails)
```
