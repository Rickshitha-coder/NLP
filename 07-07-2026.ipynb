```python
from collections import defaultdict

corpus = [
    "I want pizza",
    "I want burger",
    "We want pizza",
    "They want dosa",
    "You want burger"
]

D = 0.75

bigram = defaultdict(int)
history = defaultdict(int)

for sentence in corpus:
    words = sentence.split()

    for i in range(len(words)-1):
        bigram[(words[i], words[i+1])] += 1
        history[words[i]] += 1

continuation = defaultdict(set)

for (w1, w2) in bigram:
    continuation[w2].add(w1) #previous word
# result of previous loop example → { “want” : {I,We,They,You}, “pizza”: {want}, “burger”:{want} , “dosa”:{dosa}}
followers = defaultdict(set)

for (w1, w2) in bigram: #next word
    followers[w1].add(w2)
#result of the above loop: { “I” : {want} , “want”: {“pizza”,”burger”,”dosa”}, “we”:{want},”they”:{want},”You”:{want}}

total_unique_bigrams = len(bigram) #output → 7

def kneser_ney(prev, curr):
    count = bigram.get((prev, curr), 0) # want,pizza → 2
    history_count = history.get(prev, 0) # 5

    first = max(count - D, 0) / history_count  # 2 - 0.75 /5 → 0.25
    lamb = (D * len(followers[prev])) / history_count #0.75 * 3 / 5 → 0.45

    cont_prob = len(continuation[curr]) / total_unique_bigrams # 1/7 → 0.14

    return first + lamb * cont_prob #0.25 + 0.45 * 0.14

foods = ["pizza", "burger", "dosa"]

for food in foods:
    print(food, ":", round(kneser_ney("want", food), 3))
```
